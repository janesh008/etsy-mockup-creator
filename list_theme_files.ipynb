# Theme JPG File Name Collector & Analyzer

This notebook mounts your Google Drive, scans a main folder containing **Date Folders**, inside of which are **Theme Folders**, and collects all the `.jpg` / `.jpeg` file names. The list is then saved to a structured text file in your Drive.

### Step 2: Configure Folders
Update the paths below to match the folder locations on your Google Drive.

```python
import os

# TODO: Change this to the path of your Main folder inside Google Drive
MAIN_FOLDER_PATH = '/content/drive/MyDrive/your_main_folder_here'

# Path where you want to save the output text file containing the file names
OUTPUT_TXT_PATH = '/content/drive/MyDrive/theme_jpg_files_list.txt'
```

### Step 3: Scan JPG Files and Generate Report
Run this cell to collect all JPG file names from the subfolders and save them to the output text file.

```python
def analyze_and_collect_jpgs(main_dir, output_file):
    if not os.path.exists(main_dir):
        print(f"Error: The directory '{main_dir}' does not exist. Please check your path configuration.")
        return
    
    print(f"Scanning main directory: {main_dir}")
    
    with open(output_file, 'w', encoding='utf-8') as f:
        f.write("==================================================\n")
        f.write("Theme JPG Files Inventory Report\n")
        f.write(f"Main Folder: {main_dir}\n")
        f.write("==================================================\n\n")
        
        total_dates = 0
        total_themes = 0
        total_jpgs = 0
        
        # Get all Date folders (direct subfolders of the main directory)
        try:
            date_folders = sorted([d for d in os.listdir(main_dir) if os.path.isdir(os.path.join(main_dir, d))])
        except Exception as e:
            print(f"Error reading main directory: {e}")
            return
            
        for date_folder in date_folders:
            date_path = os.path.join(main_dir, date_folder)
            f.write(f"📅 Date Folder: {date_folder}\n")
            f.write("==================================================\n")
            total_dates += 1
            
            # Get all Theme folders inside the Date folder
            theme_folders = sorted([t for t in os.listdir(date_path) if os.path.isdir(os.path.join(date_path, t))])
            
            if not theme_folders:
                f.write("  (No theme folders found in this date folder)\n\n")
                continue
                
            for theme_folder in theme_folders:
                theme_path = os.path.join(date_path, theme_folder)
                f.write(f"  📂 Theme: {theme_folder}\n")
                f.write("  --------------------------------------------------\n")
                total_themes += 1
                
                # Collect all JPG / JPEG files inside the Theme folder
                try:
                    files = os.listdir(theme_path)
                    jpg_files = sorted([file for file in files if file.lower().endswith(('.jpg', '.jpeg'))])
                except Exception as e:
                    f.write(f"    Error reading theme folder: {e}\n\n")
                    continue
                
                if not jpg_files:
                    f.write("    (No JPG files found in this theme folder)\n")
                else:
                    for jpg in jpg_files:
                        f.write(f"    - {jpg}\n")
                        total_jpgs += 1
                f.write("\n")
            f.write("\n")
            
        f.write("==================================================\n")
        f.write("Final Summary:\n")
        f.write(f"  - Total Date Folders: {total_dates}\n")
        f.write(f"  - Total Theme Folders: {total_themes}\n")
        f.write(f"  - Total JPG Files: {total_jpgs}\n")
        f.write("==================================================\n")
        
    print(f"\nSuccess! Collected {total_jpgs} JPG files from {total_themes} themes in {total_dates} date folders.")
    print(f"Report saved to: {output_file}")

# Run the collector
analyze_and_collect_jpgs(MAIN_FOLDER_PATH, OUTPUT_TXT_PATH)
```
